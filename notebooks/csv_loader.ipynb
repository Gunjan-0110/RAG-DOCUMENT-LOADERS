```python
from langchain_community.document_loaders.csv_loader import CSVLoader
from pathlib import Path
from pprint import pp
```

```python
#define the path to the csv file
file_path=Path("../knowledge-source/organizations.csv")
file_path.exists()   #check if path exists
```

```python
#create the csv loader
loader=CSVLoader(file_path=file_path,
                 source_column="Founded",  #column to be used as source)
                 metadata_columns=["Website","Founded","Number of employees"]    #columns to be used in metadata and similarly it can be used for source as well by using content_column parameter
                 )  
```

```python
#load the document
documents=loader.load()
```

```python
#length of documents 
len(documents)  #total no.of rows in the csv file
```

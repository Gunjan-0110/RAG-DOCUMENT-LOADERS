```python
from langchain_community.document_loaders import WebBaseLoader
from pprint import pp
```

```python
#create a loader for the web page
url_1="https://docs.langchain.com/oss/python/integrations/document_loaders#webpages"
url_2="https://reference.langchain.com/python/langchain-community/document_loaders/pdf/PDFMinerLoader"
loader=WebBaseLoader(web_path=[url_1,url_2])
```

```python
#Load the document
documents=loader.load()
```

## Recursive Loader

```python
from langchain_community.document_loaders import RecursiveUrlLoader
```

```python
#create the loader for the web page
recursive_url="https://reference.langchain.com/python/langchain-community/document_loaders"
loader=RecursiveUrlLoader(url=recursive_url,
                          max_depth=2)
```

```python
#Load the document
documents=loader.load()
```

```python
len(documents)
```

```python
for i in range(10):
    doc=documents[i]
    print(doc.metadata, end="\n\n")
```

```python
print(documents[0].page_content)
```

```python
pp(documents[0].metadata)
```

```python
#lazy load
documents_lazy_load=loader.lazy_load()
```

```python
documents_lazy_load
```

```python
counter=0
 
for document in documents_lazy_load:
    #stop condition
    if counter==20:
        break
    
    #increment counter
    counter+=1
    
    #print the metadata & page_content of the document
    print(document.page_content[0:300], end="\n")
    pp(document.metadata)
```

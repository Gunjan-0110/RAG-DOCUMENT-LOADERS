```python
from langchain_community.document_loaders.pdf import PyPDFLoader
from pathlib import Path
from pprint import pp
```

```python
#create the path to the pdf file
file_path=Path("../knowledge-source/attention_is_all_you_need.pdf")
file_path.exists()
```

## PyPDF Loader


```python
file_path.as_posix()
```

```python
#create a loader
pypdf_loader=PyPDFLoader(file_path=file_path.as_posix(),mode='page')
```

```python
#load the documents
documents=pypdf_loader.load()
```

```python
len(documents)  #number of documents loaded 
```

```python
type(documents)
print(documents[0])  #fetch the first document page content
```

```python
#displaying metadata of the first page
pp(documents[0].metadata)
```

```python
from langchain_community.document_loaders.parsers.images import RapidOCRBlobParser,TesseractBlobParser
```

```python
#create pypdf instance which can create images
pypdf_image_loader = PyPDFLoader(
    file_path=file_path.as_posix(),
    mode='page',
    extract_images=True,
    images_parser=RapidOCRBlobParser(),
    images_inner_format='html-img'
)
```

```python
documents_with_images = pypdf_image_loader.load()
```

```python
#length of the documents with images
len(documents_with_images)
```

```python
page_with_image=documents_with_images[2]
print(page_with_image.page_content)  #content of the page with image
```

```python
print(page_with_image.page_content[:700])  #first 700 characters of the page content with image
```

## PDF Miner


```python
from langchain_community.document_loaders import PDFMinerLoader
```

```python
#create pdf miner loader instance
pdf_miner_loader = PDFMinerLoader(
    file_path=file_path.as_posix(),
    mode='page',
    extract_images=True,
    images_parser=RapidOCRBlobParser(),
    images_inner_format='markdown-img'
)
```

```python
#load the document
documents_miner = pdf_miner_loader.load()
```

```python
print(documents_miner[2].page_content[-500:])  #last 500 characters of the page content with image
```

```python
#table content
print(documents_miner[5].page_content)  
```

```python
pp(documents_miner[0].metadata)
```

```python
for doc in documents_miner:
    if doc.metadata["page"]==7:
        print(doc.page_content)
        break
```

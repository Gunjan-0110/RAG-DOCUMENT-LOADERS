```python
from langchain_community.document_loaders.json_loader import JSONLoader
from pathlib import Path
from pprint import pp
```

```python
#create path to the json file
file_path=Path("../knowledge-source/apparels.json")
file_path.exists()   #check if path exists
```

```python
def metadata_func(record: dict, metadata: dict)->dict:
    metadata["productName"]=record.get("productName")
    metadata["price"]=record.get("price")
    metadata["category"]=record.get("category")
    #delete seq_number 
    del metadata["seq_num"]
    return metadata
```

```python
#form the json loader
loader=JSONLoader(file_path=file_path.as_posix(),
                  jq_schema=".products[]",
                  content_key="Description",
                  metadata_func=metadata_func
                  )
```

```python
#load the document
documents=loader.load()
```
